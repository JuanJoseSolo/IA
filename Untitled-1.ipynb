```python
import matplotlib.pyplot as plt

# Define the points
x = [0, -0.1]
y = [0.4, 0]

# Extend the line beyond the x ticks
x_extended = [x[0], x[1], None, x[0]-0.5, x[1]+0.5]  # Extend by adding points before and after the original points
y_extended = [y[0], y[1], None, 0, 0]  # Extend by adding points before and after the original points

# Plot the extended line
plt.plot(x_extended, y_extended, marker='o', linestyle='-')  # marker='o' for circles, linestyle='-' for a solid line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Extended Line Beyond X Ticks')
plt.grid(True)
plt.show()
```
